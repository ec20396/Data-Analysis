```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def plot_yellow_month(yellow_month):
    # Extracting data for yellow month
    yellow_month_labels = yellow_month['pickup_month']
    yellow_trip_count = yellow_month['trip_count']

    # Plotting the bar chart with custom style for yellow data
    plt.figure(figsize=(2, 5))
    plt.bar(yellow_month_labels, yellow_trip_count, color='yellow', alpha=0.7, edgecolor='black',label='Yellow Taxi')
    plt.xlabel('Month')
    plt.ylabel('Trip Count')
    plt.gca().get_yaxis().set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.title('Taxi Trip Count by Month - Yellow Taxi')
    plt.show()

def plot_green_month(green_month):
    # Extracting data for green month
    green_month_labels = green_month['pickup_month']
    green_trip_count = green_month['trip_count']

    # Plotting the bar chart with custom style for green data
    plt.figure(figsize=(2, 5))
    plt.bar(green_month_labels, green_trip_count, color='green', alpha=0.7, edgecolor='black', label='Green Taxi')
    plt.xlabel('Month')
    plt.ylabel('Trip Count')
    plt.gca().get_yaxis().set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.title('Taxi Trip Count by Month - Green Taxi')
    plt.show()

if __name__ == "__main__":
    local_csv_path_yellow = "/home/jovyan/coursework/Task10/month_yellow.csv"
    local_csv_path_green = "/home/jovyan/coursework/Task10/month_green.csv"

    yellow_month = pd.read_csv(local_csv_path_yellow)
    green_month = pd.read_csv(local_csv_path_green)

    # Call the functions to plot both yellow and green months
    plot_yellow_month(yellow_month)
    plot_green_month(green_month)
```
