```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker

def plot_yellow_pickup_counts(yellow_pickup_counts_df):
    boroughs = yellow_pickup_counts_df['Pickup_Borough']
    yellow_pickup_counts = yellow_pickup_counts_df['Yellow_Pickup_Count']

    plt.plot(boroughs, yellow_pickup_counts, marker='o', color='yellow', label='Yellow Pickup Counts')
    plt.xticks(rotation='vertical')
    plt.xlabel("Borough")
    plt.ylabel("Pickup Counts")
    plt.title("Yellow Taxi Pickup Counts in Each NYC Borough")
    plt.gca().get_yaxis().set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task2/yellow_pickup.csv"
    yellow_pickup_counts_df = pd.read_csv(local_csv_path)




def plot_yellow_dropoff_counts(yellow_dropoff_counts_df):
    boroughs = yellow_dropoff_counts_df['Dropoff_Borough']
    yellow_dropoff_counts = yellow_dropoff_counts_df['Yellow_Dropoff_Count']

    plt.plot(boroughs, yellow_dropoff_counts, marker='o', color='yellow', label='Yellow Dropoff Counts')
    plt.xticks(rotation='vertical')
    plt.xlabel("Borough")
    plt.ylabel("Dropoff Counts")
    plt.title("Yellow Taxi Dropoff Counts in Each NYC Borough")
    plt.gca().get_yaxis().set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task2/yellow_dropoff.csv"
    yellow_dropoff_counts_df = pd.read_csv(local_csv_path)



def plot_green_dropoff_counts(green_dropoff_counts_df):
    boroughs = green_dropoff_counts_df['Dropoff_Borough']
    green_dropoff_counts = green_dropoff_counts_df['Green_Dropoff_Count']

    plt.plot(boroughs, green_dropoff_counts, marker='o', color='green', label='Green Dropoff Counts')
    plt.xticks(rotation='vertical')
    plt.xlabel("Borough")
    plt.ylabel("Dropoff Counts")
    plt.title("Green Taxi Dropoff Counts in Each NYC Borough")
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task2/green_dropoff.csv"
    green_dropoff_counts_df = pd.read_csv(local_csv_path)
    




def plot_green_pickup_counts(green_pickup_counts_df):
    boroughs = green_pickup_counts_df['Pickup_Borough']
    green_pickup_counts = green_pickup_counts_df['Green_Pickup_Count']

    plt.plot(boroughs, green_pickup_counts, marker='o', color='green', label='Green Pickup Counts')
    plt.xticks(rotation='vertical')
    plt.xlabel("Borough")
    plt.ylabel("Pickup Counts")
    plt.title("Green Taxi Pickup Counts in Each NYC Borough")
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task2/green_pickup.csv"
    green_pickup_counts_df = pd.read_csv(local_csv_path)
    
```
