```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker

def plot_daily_trip_counts(daily_trip_counts_df):
    Date = daily_trip_counts_df['Date']
    trip_counts = daily_trip_counts_df['count']

    plt.plot(Date, trip_counts, marker='o', color='red', label='Daily Trip Counts For 1st Week of 2023')
    plt.xticks(rotation='vertical')
    plt.xlabel("Date")
    plt.ylabel("Trip Counts")
    plt.title("Daily Trip Counts For 1st Week of 2023")
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task3/daily_trip_counts.csv"
    daily_trip_counts_df = pd.read_csv(local_csv_path)

```

```python
plot_daily_trip_counts(daily_trip_counts_df)
```
