```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker

def plot_top_yellow_pickup_count(top_yellow_pickup_counts_df):
    boroughs = top_yellow_pickup_counts_df['Pickup_Borough']
    top_yellow_pickup_counts = top_yellow_pickup_counts_df['count']

    plt.plot(boroughs, top_yellow_pickup_counts, marker='o', color='yellow', label='Yellow Pickup Counts')
    plt.xticks(rotation='vertical')
    plt.xlabel("Top 5 Boroughs")
    plt.ylabel("Top 5 Yellow Taxi Pickup Locations")
    plt.title("Top 5 Yellow Taxi Pickup Locations and there totals")
    plt.gca().get_yaxis().set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task4/top_yellow_pickup.csv"
    top_yellow_pickup_counts_df = pd.read_csv(local_csv_path)



def plot_top_yellow_dropoff_count(top_yellow_dropoff_counts_df):
    boroughs = top_yellow_dropoff_counts_df['Dropoff_Borough']
    top_yellow_dropoff_counts = top_yellow_dropoff_counts_df['count']

    plt.plot(boroughs, top_yellow_dropoff_counts, marker='o', color='yellow', label='Yellow Dropoff Counts')
    plt.xticks(rotation='vertical')
    plt.xlabel("Top 5 Boroughs")
    plt.ylabel("Top 5 Yellow Taxi Drop off Counts")
    plt.title("Top 5 Yellow Taxi Drop off locations and there totals")
    plt.gca().get_yaxis().set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task4/top_yellow_dropoff.csv"
    top_yellow_dropoff_counts_df = pd.read_csv(local_csv_path)




def plot_top_green_dropoff_count(top_green_dropoff_counts_df):
    boroughs = top_green_dropoff_counts_df['Dropoff_Borough']
    top_green_dropoff_counts = top_green_dropoff_counts_df['count']

    plt.plot(boroughs, top_green_dropoff_counts, marker='o', color='green', label='Green Dropoff Counts')
    plt.xticks(rotation='vertical')
    plt.xlabel("Top 5 Boroughs")
    plt.ylabel("Top 5 Green Taxi Drop off Counts")
    plt.title("Top 5 Green Taxi Drop off locations and there totals")
    plt.gca().get_yaxis().set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task4/top_green_dropoff.csv"
    top_green_dropoff_counts_df = pd.read_csv(local_csv_path)


    




def plot_top_green_pickup_count(top_green_pickup_counts_df):
    boroughs = top_green_pickup_counts_df['Pickup_Borough']
    top_green_pickup_counts = top_green_pickup_counts_df['count']

    plt.plot(boroughs, top_green_pickup_counts, marker='o', color='green', label='Green Pickup Counts')
    plt.xticks(rotation='vertical')
    plt.xlabel("Top 5 Boroughs")
    plt.ylabel("Top 5 Green Taxi Pickup Counts")
    plt.title("Top 5 Green Taxi Pickup locations and there totals")
    plt.gca().get_yaxis().set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task4/top_green_pickup.csv"
    top_green_pickup_counts_df = pd.read_csv(local_csv_path)
```
