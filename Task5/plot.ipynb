```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker

def plot_anomalies_june_2023(anomalies_june_2023_df):
    Date = anomalies_june_2023_df['pickup_day']
    total_trips = anomalies_june_2023_df['total_trips']

    plt.plot(Date, total_trips, marker='o', color='red', label='Anomalies for Total Trips in June 2023')
    plt.xticks(rotation='vertical')
    plt.xlabel("Date")
    plt.ylabel("Trip Counts")
    plt.title("Anomalies Trip Counts for Days In June 2023")
    plt.show()

if __name__ == "__main__":
    local_csv_path = "/home/jovyan/coursework/Task5/anomalies.csv"
    anomalies_june_2023_df = pd.read_csv(local_csv_path)
```

```python
plot_anomalies_june_2023(anomalies_june_2023_df)
```
